```python
import os
from pathlib import Path
from openai import OpenAI
from dotenv import load_dotenv

# 1. .env 파일에 저장된 API 키 불러오기
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# 2. 생성될 오디오 파일 경로 설정
speech_file_path = Path("speech.mp3")

# 3. OpenAI TTS API 호출
response = client.audio.speech.create(
  model="tts-1",      # 속도가 빠른 기본 모델 (고음질은 tts-1-hd)
  voice="alloy",      # 목소리 설정: alloy, echo, fable, onyx, nova, shimmer
  input="안녕하세요 성용님! 07장 오디오 에이전트 실습 환경 구축을 축하드립니다. 제 목소리가 잘 들리시나요?"
)

# 4. 파일로 저장
response.stream_to_file(speech_file_path)

print(f"🎉 성공! {os.path.abspath(speech_file_path)} 경로에 파일이 생성되었습니다.")
```

```python
# 7-2절: 음성 인식을 위한 Whisper 모델 사용
audio_file = open("speech.mp3", "rb")

transcript = client.audio.transcriptions.create(
  model="whisper-1", 
  file=audio_file
)

print("--- 받아쓰기 결과 ---")
print(transcript.text)
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# stream=True 옵션을 주는 것이 핵심입니다.
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "user", "content": "미국 주식 시장의 장점에 대해 짧게 설명해줘."}],
    stream=True 
)

print("--- 실시간 답변 시작 ---")
for chunk in response:
    # 각 조각(chunk)에서 텍스트 내용만 뽑아서 출력
    content = chunk.choices[0].delta.content
    if content:
        print(content, end="", flush=True)
```
