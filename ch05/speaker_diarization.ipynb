```python
from docx import Document

# 1. 워드 문서 객체 생성
doc = Document()

# 2. 문서 제목 추가
doc.add_heading('AI 회의록 요약 결과', 0)

# 3. GPT가 생성한 요약 내용 가져오기 (위에서 출력된 response 내용)
summary_text = response.choices[0].message.content

# 4. 내용 추가
doc.add_paragraph(summary_text)

# 5. 파일 저장 (ch05 폴더 안에 생성됩니다)
output_path = "meeting_summary.docx"
doc.save(output_path)

print(f"🎉 모든 실습 완료! '{output_path}' 파일이 생성되었습니다.")
```
