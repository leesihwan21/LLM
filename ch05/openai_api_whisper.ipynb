```python
import torch
from transformers import pipeline

# 로컬에서 실행할 Whisper 모델 지정 (성능과 속도가 좋은 turbo 버전)
model_id = "openai/whisper-large-v3-turbo"

# 음성 인식 파이프라인 생성
# device=0은 GPU 사용, GPU가 없다면 "cpu"로 자동 설정됩니다.
pipe = pipeline(
    "automatic-speech-recognition",
    model=model_id,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device="cuda:0" if torch.cuda.is_available() else "cpu",
)

print("✅ 모델 로드 완료!")
```

```python
# 1. 파일 경로 설정
audio_file_path = "audio/lsy_audio_2023_58s.mp3"

# 2. 로컬 모델로 음성 인식 실행
# generate_kwargs를 통해 출력 언어를 한국어로 강제합니다.
result = pipe(
    audio_file_path,
    generate_kwargs={"language": "korean"}
)

# 3. 결과 출력
print("--- 🎙️ 로컬 모델 변환 결과 ---")
print(result["text"])
```

```python
import torch
import librosa
from transformers import pipeline

# 1. 모델 설정 (large 대신 메모리를 적게 먹는 small 모델로 변경)
model_id = "openai/whisper-small" 

pipe = pipeline(
    "automatic-speech-recognition",
    model=model_id,
    # GPU(cuda)를 쓸 수 있으면 쓰고, 없으면 CPU를 씁니다.
    device="cuda:0" if torch.cuda.is_available() else "cpu",
)

# 2. 음성 파일 로드
audio_file_path = "audio/lsy_audio_2023_58s.mp3"
audio_data, sr = librosa.load(audio_file_path, sr=16000)

# 3. 음성 인식 실행
result = pipe(
    audio_data,
    chunk_length_s=30,
    stride_length_s=5,
    generate_kwargs={"language": "korean"},
    return_timestamps=True
)

# 4. 결과 출력
print("-" * 30)
print("🎙️ 로컬(Small 모델) 변환 결과")
print("-" * 30)
print(result["text"])
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# .env 파일에서 API 키 로드
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# 파일 경로 (58초짜리 파일)
audio_file_path = "audio/lsy_audio_2023_58s.mp3"

# 1. Whisper API 호출 (받아쓰기)
with open(audio_file_path, "rb") as audio_file:
    transcription = client.audio.transcriptions.create(
        model="whisper-1", 
        file=audio_file
    )

print("--- 🎉 58초 파일 변환 성공! ---")
print(transcription.text)
```

```python
# 위에서 얻은 transcription.text를 요약해봅시다.
raw_text = transcription.text

response = client.chat.completions.create(
    model="gpt-4o",  # 또는 gpt-3.5-turbo
    messages=[
        {"role": "system", "content": "너는 전문 요약가야. 다음 음성 변환 텍스트를 핵심 위주로 요약해줘."},
        {"role": "user", "content": f"다음 내용을 3줄 이내로 요약해줘:\n\n{raw_text}"}
    ]
)

print("-" * 30)
print("🤖 GPT의 핵심 요약")
print("-" * 30)
print(response.choices[0].message.content)
```

```python
# 요약된 내용을 변수에 담기 (위에서 출력된 내용)
summary_content = response.choices[0].message.content

# 파일 저장하기
file_path = "summary.txt"

with open(file_path, "w", encoding="utf-8") as f:
    f.write("=== GPT 음성 요약 결과 ===\n")
    f.write(summary_content)

print(f"✅ '{file_path}' 파일로 저장이 완료되었습니다!")
```
