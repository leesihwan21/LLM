```python
import os
import base64
from glob import glob
from openai import OpenAI
from dotenv import load_dotenv

# 1. API 설정
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# 2. 이미지 인코딩 함수
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

# 3. 퀴즈 출제 함수 (프롬프트 포함)
def image_quiz(image_path):
    base64_image = encode_image(image_path)
    quiz_prompt = """
    제공한 이미지를 바탕으로, 다음과 같은 양식으로 퀴즈를 만들어 주세요.
    정답은 (1)~(4) 중 하나만 해당하도록 출제하세요.
    아래는 예시입니다.
    ----- 예시 -----
    Q: 다음 이미지에 대한 설명 중 옳지 않은 것은 무엇인가요?
    - (1) 베이커리에서 사람들이 빵을 사는 모습이 담겨 있습니다.
    - (2) 맨 앞에 서 있는 사람은 빨간색 셔츠를 입었습니다.
    - (3) 기차를 타기 위해 줄을 서 있는 사람들이 있습니다.
    - (4) 점원은 노란색 티셔츠를 입었습니다.

    정답: (4) 점원은 노란색 티셔츠가 아닌 파란색 티셔츠를 입었습니다.
    (주의: 정답은 (1)~(4) 중 하나만 선택하도록 출제하세요.)
    ======
    """
    
    messages = [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": quiz_prompt},
                {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}},
            ],
        }
    ]
    
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=messages
    )
    return response.choices[0].message.content

# 4. 폴더 내 모든 이미지를 읽어서 문제집(마크다운) 생성하기
txt = ""
no = 1

# ./data/images/ 폴더 안의 모든 jpg 파일을 찾습니다.
for g in glob('./data/images/*.jpg'):
    try:
        print(f"[{no}번 문제 생성 중...] 파일: {os.path.basename(g)}")
        q = image_quiz(g)
    except Exception as e:
        print(f"에러 발생: {e}")
        continue
    
    # 마크다운 서식 조립
    divider = f'## 문제 {no}\n\n'
    txt += divider
    filename = os.path.basename(g)
    txt += f'![image]({filename})\n\n'
    
    # 화면 출력 및 텍스트 누적
    print(q)
    txt += q + '\n\n------------------------\n\n'
    
    # 마크다운 파일로 저장 (.md)
    with open('./data/images/image_quiz.md', 'w', encoding='utf-8') as f:
        f.write(txt)
        
    no += 1

print("\n🎉 완료! data/images 폴더 안에 image_quiz.md 파일이 생성되었습니다!")
```
